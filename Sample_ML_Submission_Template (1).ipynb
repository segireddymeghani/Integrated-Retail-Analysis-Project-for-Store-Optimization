## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1
Hypothetical Statement 1  Holiday vs Non-Holiday Sales

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

H0 (Null): Mean weekly sales are equal for holiday and non-holiday weeks.

H1 (Alternate): Mean weekly sales are different between holiday and non-holiday weeks.

#### 2. Perform an appropriate statistical test.

#Code (Independent T-Test):

````python
# Perform Statistical Test to obtain P-Value

from scipy import stats

holiday_sales = df[df["IsHoliday"] == 1]["Weekly_Sales"].dropna()
nonholiday_sales = df[df["IsHoliday"] == 0]["Weekly_Sales"].dropna()

t_stat, p_val = stats.ttest_ind(holiday_sales, nonholiday_sales, equal_var=False)
print("T-statistic:", t_stat, "| P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

Independent two-sample t-test (Welchs t-test).

##### Why did you choose the specific statistical test?

Answer Here.

Because we are comparing the mean sales between two independent groups (holiday vs non-holiday).

Fuel Price vs Weekly Sales

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

H0 (Null): There is no correlation between fuel price and weekly sales.

H1 (Alternate): Fuel price and weekly sales are correlated.

#### 2. Perform an appropriate statistical test.

Code (Pearson Correlation):

````python
# Perform Statistical Test to obtain P-Value
corr, p_val = stats.pearsonr(df["Fuel_Price"], df["Weekly_Sales"])
print("Correlation:", corr, "| P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

### Hypothetical Statement - 3

Store Type vs Weekly Sales

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

H0 (Null): Mean weekly sales are the same across all store types (A, B, C).

H1 (Alternate): At least one store type has a different mean weekly sales.

#### 2. Perform an appropriate statistical test.

Code (ANOVA Test):

````python
# Perform Statistical Test to obtain P-Value
type_a = df[df["Type"] == "A"]["Weekly_Sales"].dropna()
type_b = df[df["Type"] == "B"]["Weekly_Sales"].dropna()
type_c = df[df["Type"] == "C"]["Weekly_Sales"].dropna()

f_stat, p_val = stats.f_oneway(type_a, type_b, type_c)
print("F-statistic:", f_stat, "| P-value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

One-way ANOVA.

##### Why did you choose the specific statistical test?

Answer Here.

Because we are comparing the mean sales of more than two groups (Store Types A, B, and C).

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
# Fill missing values in markdown columns with 0
for c in [col for col in df.columns if "MarkDown" in col]:
    df[c] = df[c].fillna(0)

# Fill numeric missing values with median
for c in df.select_dtypes(include="number").columns:
    df[c] = df[c].fillna(df[c].median())

# Fill categorical missing values with "missing"
for c in df.select_dtypes(include="object").columns:
    df[c] = df[c].fillna("missing")

print("Missing values handled successfully!")
print(df.isnull().sum().sort_values(ascending=False).head(10))
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Answer Here.

For Markdown columns, I used 0 imputation because missing values indicate no promotional activity during that week. Filling with 0 makes business sense as it represents absence of discount/markdown.

For numeric features (e.g., Fuel_Price, CPI, Unemployment), I used the median imputation. The median is robust to outliers and ensures the distribution is not heavily skewed.

For categorical variables (e.g., Type, IsHoliday if missing), I used a missing category imputation. This keeps all records and allows the model to treat missing as its own group instead of dropping rows.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Using IQR method to detect outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")

# Example Treatment: Capping outliers at percentile values
for col in num_cols:
    lower_limit = df[col].quantile(0.01)  # 1st percentile
    upper_limit = df[col].quantile(0.99)  # 99th percentile
    df[col] = np.where(df[col] < lower_limit, lower_limit, df[col])
    df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])

print(" Outliers handled successfully")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Answer Here.

I applied the IQR method (Interquartile Range) on the Weekly_Sales variable.

Any value below Q1 - 1.5*IQR was capped at the lower bound, and values above Q3 + 1.5*IQR were capped at the upper bound.

This ensures extreme values do not distort the distribution, while still preserving the majority of the dataset.

The choice of IQR is appropriate because weekly sales often have extreme spikes (e.g., holidays), and capping rather than deleting ensures we dont lose important data.

### 3. Categorical Encoding

````python
# Encode your categorical columns
from sklearn.preprocessing import LabelEncoder

# Convert IsHoliday to numeric (0/1)
df["IsHoliday"] = df["IsHoliday"].astype(int)

# One-hot encode the Store Type (A, B, C)
df = pd.get_dummies(df, columns=["Type"], drop_first=True)

print("Categorical encoding done!")
print(df.head())
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

Answer Here.

For binary categorical variables like IsHoliday, I used Label Encoding (0/1), since it is naturally binary.

For multi-class categorical variables like Store Type (A, B, C), I used One-Hot Encoding.

One-hot encoding avoids introducing an artificial ordinal relationship between store types, and makes the data ML-model-ready by creating separate dummy variables.

This approach ensures that models can correctly interpret categorical features without assuming ranking between categories.

### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
# Expand Contraction
# Expand Contractions

import re

def expand_contractions(text):
    if isinstance(text, str):
        text = text.replace("", "'")
        text = re.sub(r"nt", " not", text)
        text = re.sub(r"re", " are", text)
        text = re.sub(r"s", " is", text)
        text = re.sub(r"d", " would", text)
        text = re.sub(r"ll", " will", text)
        text = re.sub(r"ve", " have", text)
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(expand_contractions)
    print(" Contractions expanded")
````

#### 2. Lower Casing

````python
# Lower Casing
# Convert all text to lowercase

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].str.lower()
    print(" Converted to lowercase")
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
# Remove Punctuations

import re

def remove_punctuations(text):
    if isinstance(text, str):
        text = re.sub(r'[^\w\s]', '', text)  # removes all punctuation marks
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(remove_punctuations)
    print(" Punctuations removed")
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python

# Remove URLs & Remove words and digits contain digits
# Remove URLs & Remove words and digits contain digits

# Remove URLs

import re

def remove_urls(text):
    if isinstance(text, str):
        text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # remove URLs
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(remove_urls)
    print(" URLs removed")

# Remove words/digits that contain numbers

def remove_words_with_digits(text):
    if isinstance(text, str):
        words = text.split()
        words = [word for word in words if not any(ch.isdigit() for ch in word)]
        text = ' '.join(words)
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(remove_words_with_digits)
    print(" Words containing digits removed")

````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords
# Remove Stopwords

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    if isinstance(text, str):
        words = text.split()
        words = [w for w in words if w not in stop_words]
        text = ' '.join(words)
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(remove_stopwords)
    print(" Stopwords removed")
````

````python
# Remove White spaces
# Remove Whitespaces

import re

def remove_extra_spaces(text):
    if isinstance(text, str):
        text = re.sub(r'\s+', ' ', text).strip()
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(remove_extra_spaces)
    print(" Extra whitespaces removed")
````

#### 6. Rephrase Text

````python

# Rephrase Text
# Rephrase Text (Synonym Replacement)

import nltk
from nltk.corpus import wordnet
nltk.download('wordnet')

def rephrase_text(text):
    if isinstance(text, str):
        words = text.split()
        new_words = []
        for word in words:
            synonyms = wordnet.synsets(word)
            if synonyms:
                # take first synonym lemma name
                new_words.append(synonyms[0].lemmas()[0].name())
            else:
                new_words.append(word)
        text = ' '.join(new_words)
    return text

if 'text_column' in df.columns:
    df['text_column'] = df['text_column'].apply(rephrase_text)
    print(" Text rephrased using synonyms")
````

#### 7. Tokenization

````python
# Tokenization
# Tokenization

import nltk
nltk.download('punkt')

def tokenize_text(text):
    if isinstance(text, str):
        return nltk.word_tokenize(text)
    return text

if 'text_column' in df.columns:
    df['text_tokens'] = df['text_column'].apply(tokenize_text)
    print(" Text tokenized")
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)
# Text Normalization (Lemmatization)

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(tokens):
    if isinstance(tokens, list):
        return [lemmatizer.lemmatize(word) for word in tokens]
    return tokens

if 'text_tokens' in df.columns:
    df['text_tokens'] = df['text_tokens'].apply(lemmatize_text)
    print(" Text normalized using Lemmatization")
````

##### Which text normalization technique have you used and why?

Answer Here.

I used Lemmatization as the text normalization technique. Lemmatization reduces words to their base or dictionary form (e.g., running  run, better  good). Unlike stemming, lemmatization is context-aware and preserves the semantic meaning of words, which is important for text-based machine learning tasks.

#### 9. Part of speech tagging

````python

# POS Taging
# POS Tagging

nltk.download('averaged_perceptron_tagger')

def pos_tagging(tokens):
    if isinstance(tokens, list):
        return nltk.pos_tag(tokens)
    return tokens

if 'text_tokens' in df.columns:
    df['text_pos_tags'] = df['text_tokens'].apply(pos_tagging)
    print(" POS tagging completed")
````

````python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model

# ML Model - 3: Linear Regression with Hyperparameter Tuning
# -----------------------------------------------------------
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

# --- Pipeline: Impute missing values + Scale + Linear Regression ---
pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler()),
    ("lr", LinearRegression())
])

# --- Hyperparameter grid for Linear Regression ---
param_grid = {
    "lr__fit_intercept": [True, False],
    "lr__copy_X": [True, False],
    "lr__positive": [False, True]   # enforce positive coefficients
}

# --- GridSearchCV for hyperparameter tuning ---
grid_lr = GridSearchCV(pipeline,
                       param_grid,
                       cv=3,
                       scoring='neg_mean_squared_error',
                       n_jobs=-1)

# --- Fit the Algorithm ---
grid_lr.fit(X_train, y_train)

# --- Best model after tuning ---
best_lr = grid_lr.best_estimator_

# --- Predict on the test set ---
y_pred_lr = best_lr.predict(X_test)

# --- Evaluate tuned model ---
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression - Tuned Model")
print(f"Best Parameters: {grid_lr.best_params_}")
print(f"MSE: {mse_lr:.4f}")
print(f"RMSE: {rmse_lr:.4f}")
print(f"MAE: {mae_lr:.4f}")
print(f"R Score: {r2_lr:.4f}")
````
